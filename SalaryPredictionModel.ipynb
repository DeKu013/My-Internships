```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Salary_dataset.csv")
```

```python
df
```

```python
#Spilling the data into dependent data and independent data
```

```python
x = df[['YearsExperience']]               #Independent data
y = df[['Salary']]                        #Dependent data
```

```python
#Data visuaization
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Years of experience VS Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(x,y,color='green',s=50)#,ec=['yellow','orange','black','red','blue'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
model = LR()
```

```python
model.fit(x,y)
```

```python
#Simple Regression: y = m*x + c
```

```python
model.predict([[2]])
```

```python
model.coef_              #Value of 'm'
```

```python
model.intercept_         #Value of 'c'
```

```python
model.coef_ * 2 + model.intercept_             #Model verification
```

```python
pred = model.predict(x)
```

```python
plt.title('Years of experience VS Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.scatter(x,y)            #Actual data
plt.scatter(x,pred,color='orange')         #Predicted data
```

```python
plt.title('Years of experience VS Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.scatter(x,y)            #Actual data
plt.plot(x,pred,color='green')         #Predicted data
```

```python
#Interface to take inputs
```

```python
myexp = int(input('Enter the years of experience:'))
mysal = model.predict([[myexp]])
newsal = mysal[0]
newsalrd = np.round(newsal,2)
print('The salary is:',*newsalrd)
```
